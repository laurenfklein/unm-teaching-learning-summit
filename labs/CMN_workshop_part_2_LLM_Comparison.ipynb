## Filling in the blank, redux

Below you can find some code that does the very same thing we asked our OLMo model to do via the web interface.

As above, **hover on the cell and press the black arrow button to run the code.**

```
# this is how you specify the model you want to use
unmasker = pipeline('fill-mask',
                     model = 'distilroberta-base',  # this is the name of the model
                     )

# this is where you specify the prompt
prompt = "After college, she got a job as a <mask>"

# this is how you actually query the model
responses = unmasker(prompt,
                    top_k = 10, # number of responses
                      )
# print out the responses in a nice way
print("Responses to the following prompt:\n")
print(prompt + "\n")

print(f"{"Word":<15} {"Score":>15}")
print(f"{"-"*15:<15} {"-"*15:>15}")
for response in responses:
  print(f"{response["token_str"].strip():<15} {response["score"]:>15.3f}")
```

## The same prompt, ninteenth-century style!

We've been talking about all the problems with LLMs that can be traced to the fact that the training data comes from the internet. But as it turns out, there are researchers who are training LLMs using only historical data. In fact, that's how the [BLERT](https://huggingface.co/Livingwithmachines/bert_1760_1900) model was trained: using a giant number of digitized newspapers and other texts from the British Library that span the years 1760 to 1900. Now we're going to use it for the same prompt.

```
unmasker = pipeline('fill-mask',
                     model = 'Livingwithmachines/bert_1760_1850',  # note different model
                     )

# same prompt as above!
prompt = "On Saturday, he went to the [MASK]"

# same way of formatting the responses
responses = unmasker(prompt,
                    top_k = 10, # number of responses
                      )

# same way of printing out the responses
print("Responses to the following prompt:\n")
print(prompt + "\n")

print(f"{"Word":<15} {"Score":>15}")
print(f"{"-"*15:<15} {"-"*15:>15}")
for response in responses:
  print(f"{response["token_str"].strip():<15} {response["score"]:>15.3f}")
```

The takeaway, here, is that *training data matters.*  Thinking back to Donna Haraway, we might say that training data (or the model) is *situated*. Or we might recall the quote from Ngũgĩ wa Thiong'o's *Decolonizing the Mind*:

> “Language carries culture, and culture carries... the entire body of values by which we come to perceive ourselves and our place in the world. How people perceive themselves affects how they look at their culture, at their places, their politics and at the social production of wealth, at their entire relationship to nature and to other beings. Language is thus inseparable from ourselves as a community of human beings with a specific form and character, a specific history, a specific relationship to the world.”

> — Ngũgĩ wa Thiong'o, Decolonising the Mind: The Politics of Language in African Literature (1986), p. 16.
